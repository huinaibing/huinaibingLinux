```python
import numpy as np
from matplotlib import pyplot as plt


def mcmc_sampling(f, n_samples, step=0.1):
    """
    
    :param f: 
    :param n_samples: 
    :param step: 
    :return: 
    """
    samples = []
    x0 = 0.0
    samples.append(x0)
    for i in range(n_samples - 1):
        x1 = x0 + step * np.random.uniform(low=-1., high=1.)
        
        alpha = min(1.0, f(x1) / f(x0))
        r = np.random.rand()
        if r < alpha:
            x0 = x1
        samples.append(x0)
    
    return samples
```

```python
def f(x):
    return np.exp(-np.abs(x))
```

```python
import numpy as np
from matplotlib import pyplot as plt


def mcmc_sampling_nd(f, n_samples, dim, step=0.1):
    """
    
    :param f: 
    :param n_samples: 
    :param dim: 
    :param step: 
    :return: 
    """
    samples = []
    x0 = np.zeros(dim)
    samples.append(x0)
    for i in range(n_samples - 1):
        x1 = x0 + step * np.random.uniform(low=-1., high=1., size=dim)
        
        alpha = min(1.0, f(x1) / f(x0))
        r = np.random.rand()
        if r < alpha:
            x0 = x1
        samples.append(x0)
    
    return samples
```

```python
def f_2d(x, y):
    return np.exp(-np.abs(x)) * np.exp(-np.abs(y))
```

```python
import numpy as np
import matplotlib.pyplot as plt
import time
from IPython.display import clear_output
def f(x1, x2):
    return 0.1 * x1**2 + 2 * x2**2
def f_grad(x1, x2):
    return 0.2 * x1, 4 * x2
def visualize(track):
    x1 = np.linspace(-5, 3, 100)
    x2 = np.linspace(-3, 3, 100)
    
    xx1, xx2 = np.meshgrid(x1, x2)
    plt.contour(xx1, xx2, f(xx1, xx2))
    plt.colorbar()
    plt.plot([0], [0], "ro")
    track = np.array(track)
    plt.plot(track[:, 0], track[:, 1], "bo-")
    plt.show()
    # print(track)
```

```python
n_iter = 5000
track = []
x10, x20 = -4, 2
al = 0.5
track.append((x10, x20))

# adam
m = 0
n = 0
beta1 = 0.9
beta2 = 0.9

eps = 0.001

for i in range(n_iter):
    visualize(track)
    time.sleep(0.1)
    clear_output()
    g1, g2 = f_grad(x10, x20)
    g = np.array([g1, g2])
    
    #print(g)
    
    m = beta1 * m + (1 - beta1) * g
    n = beta2 * n + (1 - beta2) * g**2
    
    mhat = m / (1 - beta1**(i + 1))
    nhat = n / (1 - beta2**(i + 1))
    
    #print(nhat)
    
    x10 -= al * mhat[0] / (eps + np.sqrt(nhat[0]))
    x20 -= al * mhat[1] / (eps + np.sqrt(nhat[1]))

    track.append((x10, x20))
```
